### 1. Introducción

Este informe presenta un análisis exploratorio de datos centrado en la calidad del aire en Gijón durante el año 2022, con el objetivo específico de identificar el barrio con el aire más contaminado. La calidad del aire es esencial para la salud pública, y entender las variaciones a nivel local permite tomar medidas más efectivas para mejorar la calidad de vida de los residentes.

### 2. Objetivos del Análisis

- Identificar patrones de contaminación a lo largo del año.

- Evaluar las concentraciones de contaminantes en diferentes barrios.

- Determinar el barrio con las concentraciones más altas de contaminantes atmosféricos.

### 3. Principales contaminantes atmosféricos

> El dióxido de azufre (SO2) es un gas incoloro, con un olor penetrante cuando se encuentra en concentraciones muy elevadas. La principal fuente del SO2 es la quema de combustibles fósiles (carbón y derivados del petróleo), usados tanto para la calefacción doméstica como para la generación de electricidad y los vehículos a motor, así como en la fundición de minerales que contienen azufre. El valor máximo diario es 125 µg/m3.

>El dióxido de nitrógeno (NO2) y Oxido Nítrico (NO) son gases de color marrón y tienen un olor fuerte.El término NOX se refiere a la combinación de NO y NO2, su principal fuente es la combustión, tanto de tipo móvil (tráfico terrestre, aéreo y marítimo) como de tipo estacionario (industriales). el valor máximo anual es 40 μg/m3 .

>El Monóxido de carbono es un gas incoloro (CO), sin olor y sin gusto. Es el gas emitido más abundante, después del
CO2 y el vapor de agua. Se emite a la atmósfera por dos vías: la emisión directa y la formación a partir de otros contaminantes. La emisión directa se genera en combustiones incompletas (gas, carbón, gasóleo o biomasa), principalmente en fuentes como el tráfico y las estufas para uso doméstico. El valor máximo diario es 10 mg/m3.

>El ozono (O3) es un gas incoloro, invisible y de olor agradable, que tiene un gran poder oxidante. El ozono troposférico se encuentra en las capas bajas de la atmósfera y es considerado como contaminante. No existen fuentes destacables de O3 troposférico de origen antropogénico, sino que se trata de un contaminante secundario, que se forma a partir de otros compuestos llamados precursores, entre los que destacan los óxidos de nitrógeno (NOx) y los compuestos orgánicos volátiles (COV), que reaccionan gracias a la radiación solar. El varlo máximo diario es 120 μg/m3. 

>Partículas en suspensión (PM10 y PM2,5), a diferencia de los gases, que están formados por moléculas de una sola especie, el material particulado es una mezcla compleja de partículas sólidas y líquidas formada por un conjunto de moléculas de la misma sustancia o diferentes. Según su diámetro aerodinámico se clasifican en PM10 (diámetros inferiores a 10 micras) y PM2,5 (diámetros inferiores a 2,5 micras). Las primeras se emiten directamente a la atmósfera tanto por fenómenos naturales (incendios forestales, emisiones volcánicas) como por actividades humanas (actividades industriales, labores agrícolas, de construcción, etc.) En zonas costeras una fuente importante es el aerosol marino, formado por partículas de sal por evaporación de agua de mar. Las partículas más finas proceden principalmente de fuentes de combustión (industrias, tráfico, calderas domésticas) y también se forman por procesos químicos en la atmósferas a partir de contaminantes gaseosos (aerosol secundario). Su afectación a la salud humana depende de su composición y concentración, las PM10 pueden penetrar hasta las vías respiratorias bajas, las PM2,5
pueden penetrar hasta las zonas de intercambio de gases del pulmón. El valor máximo anual del PM10 es 40 μg/m3 y el valo máximo anual del PM2,5 es 20 μg/m3.
 
### 4. Conjunto de Datos

Los datos utilizados provienen de las estaciones de monitoreo de calidad del aire distribuidas en Gijón durante 2022, obtenidas a través de <https://datos.gob.es/es/>. Incluyen mediciones diarias de contaminantes como dióxido de azufre (SO2), dióxido de nitrógeno (NO2), partículas en suspensión (PM10, PM2.5), entre otros. Las mediciones están desglosadas por barrio en los notebooks con su nombre por estación y otro notebook con el conjunto de las 3 estaciones.

### 5. Preparación de Datos

Se realizó una limpieza de datos para abordar cualquier valor atípico, duplicado o faltante. Solo se eliminaron columnas con datos inferiores al 50% con los agentes menos contaminantes y los valores NaN se matuvieron para saber el motivo de su ausencia, que podría ser que por la baja concentración de gases no llegan al límite mínimo de medicción de la estación.

### 6. Análisis Descriptivo

Se calcularon estadísticas descriptivas para cada contaminante en cada barrio. La visualización de la distribución mediante histogramas permitió identificar posibles diferencias entre los barrios comprobando que el barrio con la estacion 1 destacaba por encima de los otros dos. La utilización de boxplots para determinar los outlyers y averiguar el motivo de por qué se produjeron teniendo en cuanta el valor máximo y su fecha de recogida. También se utilizaron jointplot de diferentes variables en los distintos barrios para comparar la dispersión de ambas variables, observando que las diferentes variables comparadas en las distintas estaciones no suelen estar muy dispersas en las estaciones 1 y 2, manteniéndose los valores cercanos al 0 y en la estación 4 sí se puede observar una mayor dispersión entre las variables a lo largo del gráfico. Por último, se realizó un análisis multivariante mediante heatmaps para comprobar la correlación de los distintos gases entre sí, destacando, en general que las variables no tienen mucha correlación, únicamente destacar la correlación por encima de 0,5 del NO y el NO2, y del las partículas PM10 y PM2,5.

### 7. Análisis Temporal

Se exploraron tendencias temporales para comprender cómo cambian las concentraciones de contaminantes a lo largo del año. Se identificaron períodos críticos al analizar el valor máximo de cada gas y la fecha en la que se produjo para, más adelante, averiguar, si se pudiera, el motivo de esos máximos.

### 8. Resultados y Conclusiones

Los resultados indican que la Estación 1 parece ser la más contaminada, presenta las concentraciones más altas de forma general en algunos contaminantes en comparación con otros barrios de Gijón. Las razones detrás de estas concentraciones elevadas deben investigarse más a fondo pero al ser una estación situada en la entrada principal a la cuidad y dependiendo del clima, su cercanía a varias estaciones industriales, se podrían relacionar así los altos niveles. No obstante, en función de las partículas sólidas PM10, la Estación 2 parece tener el aire más contaminado que a pesar de encontrarse al inicio de un gran parque, también se trata de una zona con bastante población y tránsito de vehículos.
